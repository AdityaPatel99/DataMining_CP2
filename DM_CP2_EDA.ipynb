This plot show the what age people are drinking.By observation 40 to 50 year old people mostly driking.

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x='age',hue='DRK_YN',multiple='stack')
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data=df,x='age',hue='SMK_stat_type_cd',multiple='stack')
```
